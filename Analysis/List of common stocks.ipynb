```python
file_m = open('MOEX.csv', errors='replace')
file_spb = open('SPBExchange.csv', errors='replace', encoding='utf-8')
```

```python
moex_f = pd.read_csv(file_m, header=0, index_col=0, sep=";", na_values = ["na", "", "#"])
spb_f = pd.read_csv(file_spb, header=0, index_col=0, sep=";", na_values = ["na", "", "#"])
```

## Creating a list of common stocks and making a file with them

```python
listWithCodes = [value for value in spb_f['s_RTS_code'].tolist() if value in moex_f['TRADE_CODE'].tolist()]
```

```python
new_spb = spb_f[['s_RTS_code', 'e_full_name']]
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
# new_spb.set_option('display.max_colwidth', None)
listOfStocks = new_spb.loc[new_spb['s_RTS_code'].isin(listWithCodes)]
    
```

```python
listOfStocks.to_excel('./listOfStocks.xlsx')
```
